```python
def FCFS(all_jobs,n):
	# ------------ here the algorithm for fcfs ------------------------
	fcfs=all_jobs[:]
	lst=[] #here list contains 3 values : job number , arrival time , cpu time
	Time=0
	waiting=[0]*(n+1)
	turnaround=[0]*(n+1)
	for i in fcfs:
		if(Time<i[1]):
			lst.append([-1,Time,i[1]-Time])
		Time=max(i[1],Time)
		lst.append([i[0],Time,i[2]])
		waiting [i[0]]=Time+i[2]-i[1]
		turnaround[i[0]]=Time+i[1]
		Time+=i[2]

	return lst,waiting,turnaround

def  SJF(all_jobs,n):

	sjf=[]
	for i in all_jobs :
		sjf.append(list(i))


	arr=[]
	for i in all_jobs :
		arr.append(list(i))


	jobs=[ [] for i in range(n+1)]
	cnt=0
	Time=0
	lst=[]
	waiting=[0]*(n+1)
	turnaround=[0]*(n+1)
	
	
	
	while(cnt<n):
		
		
		p=-1
		s=100000000000000
		j=0
		for i in sjf:
			#print(i)
			if(i[1]<=Time):
				
				if(i[2]>0 and i[2]<s):
					p=j
					s=i[2]
			j=j+1
		Time+=1
		
		if(p==-1):
			lst.append(-1)
		else :
			sjf[p][2]-=1
			if(sjf[p][2]==0):
				cnt+=1
				waiting[sjf[p][0]]=Time-arr[p][1]-arr[p][2]
				turnaround[sjf[p][0]]=Time-arr[p][1]
			lst.append(sjf[p][0])


	return lst,waiting,turnaround







no=int(input("enter no of jobs : "))
jobs=[ i+1 for i in range(no)]
burst=list(map(int,input("enter all list of burst space seperated : ").strip().split()))
arrival=list(map(int,input("enter all list of arrival space seperated : ").strip().split()))
all_jobs=[ [jobs[i],arrival[i],burst[i]] for i in range(no)]
dict3={}
dict4={}
for i in range(no):
    dict3[chr(ord('A')+i)]=burst[i]
for i in range(no):
    dict4[chr(ord('A')+i)]=arrival[i]
all_jobs=sorted(all_jobs,key = lambda l : l[1])


f1,f2,f3=FCFS(all_jobs,no)


i1=f1[len(f1)-1][1]+f1[len(f1)-1][2]
mp=[0]*i1
l=0
for i in f1:
    for mm in range(i[1],i[1]+i[2]):
        if i[0]==-1:
            mp[l]=0
        else:    
            mp[l]=i[0]
        l+=1
print(mp)


f11,f22,f33=SJF(all_jobs,no)
print(f11)
```
